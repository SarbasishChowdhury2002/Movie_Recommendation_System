**Title of the Project: Movie Recommendation System**

**Objective:** Movie Recommendation System is a system that seeks to predict or filter preferences according to user's choice movies. It produces a list of recommendations in any of the two ways-
Collaborating filtering: It builds model from the user's past behaviour as well as similar decision made by other users. This model is then used to predict items that users may have an interest in.
Content-based filtering: It uses a series of discrete characteristics of an item in order to recommend additional items with similar properties.
Lets develop a basic Movie Recommendation System using Python and Pandas. It just tells what movies are most similar with the user's movie choice.

**Data Source:** https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv

**Code/Implementation:**

```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Import Data

df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
# Get Features Selection

df_features = df[['Movie_Title', 'Movie_Genre', 'Movie_Language', 'Movie_Production_House', 'Movie_Crew']]
```

```python
df_features.shape
```

```python
df_features
```

```python
X = df['Movie_Title'] + ' ' + df['Movie_Genre'] + ' ' + df['Movie_Language'] + ' ' + df['Movie_Production_House'] + ' ' + df['Movie_Crew']
```

```python
X
```

```python
X.shape
```

```python
# GET FEATURE TEXT CONVERSION TO TOKENS

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

```python
# GET SIMILARITY SCORE USING COSINE SIMILARITY

# cosine_similarity computes the L2-normalized dot product of vectors.
# Euclidean (L2) normalization projects the vectors onto the unit sphere, and their dot product is then the cosine of the angle between the points denoted by the vectors.

from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similiarity_Score = cosine_similarity(X)
```

```python
Similiarity_Score
```

```python
Similiarity_Score.shape
```

```python
# GET MOVIE NAME AS INPUT FROM USER AND VALIDATE FOR CLOSEST SPELLING

movie = input("Enter your favourite movie name: ")
```

```python
all_movies_list = df['Movie_Title'].to_list()
```

```python
import difflib
```

```python
movie_recommendation = difflib.get_close_matches(movie, all_movies_list)
print(movie_recommendation)
```

```python
close_match = movie_recommendation[0]
print(close_match)
```

```python
index_of_close_match_movie = df[df.Movie_Title == close_match]['Movie_ID'].values[0]
print(index_of_close_match_movie)
```

```python
# Getting a List of Similar Movies

recommendation_score = list(enumerate(Similiarity_Score[index_of_close_match_movie]))
print(recommendation_score)
```

```python
len(recommendation_score)
```

```python
# GET ALL MOVIES SORT BASED ON RECOMMENDATION SCORE WITH RESPECT TO FAVOURITE MOVIE

sorted_similar_movies = sorted(recommendation_score, key = lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print("Top 30 movies suggested for you:")
i = 1
for m in sorted_similar_movies:
  index = m[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if i<31:
    print(i, '.', title_from_index)
    i+=1
```

```python
# TOP 10 MOVIE RECOMMENDATION SYSTEM

Movie_Name = input("Enter movie name: ")
List_of_All_Titles = df['Movie_Title'].to_list()
Find_Close_Match = difflib.get_close_matches(Movie_Name, List_of_All_Titles)
Close_Match = Find_Close_Match[0]
Index_of_Movie = df[df.Movie_Title==Close_Match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similiarity_Score[Index_of_Movie]))
Sorted_Similar_Movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse=True)
print("Top 10 Movies suggested for you:\n")
i = 1
for item in Sorted_Similar_Movies:
  index = item[0]
  title_from_index = df[df.Movie_ID==index]['Movie_Title'].values
  if i<11:
    print(i, '.', title_from_index)
    i += 1
```
